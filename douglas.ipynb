\begin{align}
toekomstige\ waarde = huidig\ waarde * rente ^{jaren}
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt

snoeikosten = 10
rentepercentage = 3

# genereer array van jaren van 0 tot 100 met stappen van 5
t = np.arange(0.0, 100.0, 5)
# maak plot met jaren tegen de toekomstige waarde van de investering
plt.plot(t, snoeikosten*(1+(rentepercentage-1)/100)**t, label='2% rente')
plt.plot(t, snoeikosten*(1+rentepercentage/100)**t, label='3% rente')
plt.plot(t, snoeikosten*(1+(rentepercentage+1)/100)**t, label='4% rente')
plt.ylabel('Waarde investering (euro)')
plt.xlabel('Tijd (j)')
plt.legend()
plt.show()
```
